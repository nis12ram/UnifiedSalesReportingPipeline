```sql
-- EACH REGION'S LAST BUSINESS DAY'S TOTAL REVENUE 
SELECT 
  SaleDate,
  Region,
  ROUND(SUM(Units * UnitPrice) / 1e6, 3) AS TotalRevenueInMillion
FROM hive_metastore.unifiedsales.UnifiedSalesDeltaTable 
WHERE SaleDate = current_date() - 2
GROUP BY SaleDate, Region
ORDER BY TotalRevenueInMillion DESC
```

```sql
--  TOP 5 MOST SOLD PRODUCT IN EACH REGION ON LAST BUSINESS DAY
WITH ProductSoldInfoPerRegion AS (
  SELECT 
    Product,
    Region,
    SUM(Units) AS TotalUnitsSold
  FROM hive_metastore.unifiedsales.UnifiedSalesDeltaTable 
  WHERE SaleDate = current_date() - 2
  GROUP BY Product, Region
  ORDER BY TotalUnitsSold DESC
)

SELECT Rank, Product, Region, TotalUnitsSold
FROM (
  SELECT 
    Product, 
    Region,
    TotalUnitsSold,
    DENSE_RANK() OVER(
      PARTITION BY region
      ORDER BY TotalUnitsSold DESC
    ) AS Rank
  FROM ProductSoldInfoPerRegion
)
WHERE Rank <= 5;
```

```sql
--  TOP 5 MOST SOLD PRODUCT ACROSS ALL REGIONS ON LAST BUSINESS DAY
WITH ProductSoldInfo AS (
  SELECT 
    Product,
    SUM(Units) AS TotalUnitsSold
  FROM hive_metastore.unifiedsales.UnifiedSalesDeltaTable 
  WHERE SaleDate = current_date() - 2
  GROUP BY Product
  ORDER BY TotalUnitsSold DESC
)
SELECT Rank, Product, TotalUnitsSold
FROM (
  SELECT 
    Product,
    TotalUnitsSold,
    DENSE_RANK() OVER(
      ORDER BY TotalUnitsSold DESC
    ) AS Rank
  FROM ProductSoldInfo
)
WHERE Rank <= 5;
```
